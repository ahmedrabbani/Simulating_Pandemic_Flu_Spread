# Original

```python
from numpy import zeros, linspace
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

Avg_num_contact = 15
beta = 0.02*Avg_num_contact 
gamma = 0.1
dt = 1             
D = 60          # Simulate for D days

t = np.linspace(0, D+1,D+1)

S = zeros(D+1)
I = zeros(D+1)
R = zeros(D+1)

# Initial condition
S[0] = 20
I[0] = 1
R[0] = 0

# Step equations forward in time
for n in range(D):
    S[n+1] = S[n] - dt*beta*S[n]*I[n]/S[0]
    I[n+1] = I[n] + dt*beta*S[n]*I[n]/S[0] - dt*gamma*I[n]
    R[n+1] = R[n] + dt*gamma*I[n]

fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number Infected')
ax.set_ylim(0,25)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
plt.title("Deterministic model - avg_num_contact = {}".format(Avg_num_contact))
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```
